```python
from PIL import Image 
from tensorflow.keras.models import load_model 
import pandas as pd
import numpy as np

model = load_model('/path/to/model_test1_version73.h5')

hue_list = []
directory = '/kaggle/input/early-detection-of-3d-printing-issues'

df_test = pd.read_csv(directory + '/test.csv')

for index in range(df_test['img_path'].count()):
    image = Image.open(directory + '/images/' + df_test['img_path'].iloc[index])
    image = image.resize((255,255)) # or (480, 640), if validation 4 
    img_arr = np.asarray(image)
    img_arr = np.expand_dims(img_arr, axis = 0)
    value = model.predict(img_arr)
    hue_list.append(round(value[0][0]))
    
# generate .csv 
df_test['has_under_extrusion'] = hue_list
df_final = df_test[['img_path','has_under_extrusion']]
df_final.to_csv('/path/to/test1.csv', index = False) # or validation_test1/4
```
