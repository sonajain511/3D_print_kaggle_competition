```python
import pandas as pd 

# reading in .csv files
df_test1 = pd.read_csv('/path/to/test1.csv')
df_val_test1_2 = pd.read_csv('/path/to/validation_test1 (2).csv')
df_val_test4_1 = pd.read_csv('/path/to/validation_test4 (1).csv')
```

```python
# initializing img_path column into the df_combo dataframe
df_combo = df_test1['img_path']

# averaging has_under_extrusion labels across the 3 csv files and rounding to next integer (0 or 1)
ave = []
for index in range(df_test1['has_under_extrusion'].count()):
    summation = df_test1['has_under_extrusion'].iloc[index] + \
                df_val_test1_2['has_under_extrusion'].iloc[index] + \
                df_val_test4_1['has_under_extrusion'].iloc[index]
    ave.append(int(round(summation/3,0)))       
```

```python
# add the has_under_extrusion column to df_combo and export
df_ave = pd.DataFrame({'has_under_extrusion':ave})
df_combo = pd.concat([df_combo,df_ave], axis = 1)
df_combo.to_csv('/path/to/combo.csv', index = False)
```
